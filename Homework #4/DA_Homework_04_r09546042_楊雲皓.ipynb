```python
import pandas as pd 
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm
```

```python
#read the dataset
data = pd.read_csv(r"C:\Users\TerryYang\pythonwork\pythonwork\Data Analytics Homework\DA_Demo.csv")
#variables column
X = data[['origin', 'model year', 'acceleration', 'weight', 'horsepower', 'displacement', 'cylinders']]
#add constant
X = sm.add_constant(X)
# VIF dataframe
vif_data = pd.DataFrame()
vif_data["Variables"] = X.columns
# calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))]
print(vif_data)
```
